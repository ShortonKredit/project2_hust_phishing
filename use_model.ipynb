```python
# ===================== Mount Google Drive =====================
from google.colab import drive
drive.mount('/content/drive')

# ===================== Import thư viện =====================
import pandas as pd
import joblib

# ===================== Load mô hình đã huấn luyện =====================
model_path = "/content/drive/MyDrive/Colab Notebooks/phishing_link_linear_model.pkl"
model = joblib.load(model_path)

# ===================== Đọc dataset mới =====================
input_csv_path = "/content/drive/MyDrive/Colab Notebooks/html_mapping_1.csv"
df = pd.read_csv(input_csv_path)

# ===================== Dự đoán nhãn và xác suất =====================
df['predicted_label'] = model.predict(df['url'])

# Lấy xác suất dự đoán
probas = model.predict_proba(df['url'])

# Thêm xác suất cao nhất vào một cột mới
df['confidence'] = probas.max(axis=1)

# ===================== Lưu kết quả ra Google Drive =====================
output_csv_path = "/content/drive/MyDrive/Colab Notebooks/html_mapping_with_labels_and_proba_1.csv"
df.to_csv(output_csv_path, index=False)

print("✅ Gán nhãn + xác suất thành công. Đã lưu vào:")
print(output_csv_path)
```
