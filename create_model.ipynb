```python
# ===================== Mount Google Drive =====================
from google.colab import drive
drive.mount('/content/drive')

# ===================== Import thư viện =====================
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import joblib

# Bước 1: Đọc dữ liệu đã làm sạch
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/combined_labeled_urls.csv")  # Đặt đúng đường dẫn nếu cần

# Bước 2: Tách dữ liệu
X = df['url']
y = df['label']

# Bước 3: Tạo pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(analyzer='char_wb', ngram_range=(3, 5))),
    ('clf', LogisticRegression(max_iter=1000, random_state=42))
])

# Bước 4: Huấn luyện mô hình
pipeline.fit(X, y)

# Bước 5: Lưu mô hình
joblib.dump(pipeline, "/content/drive/MyDrive/Colab Notebooks/phishing_link_linear_model.pkl")
print("✅ Mô hình đã được lưu thành công vào phishing_link_linear_model.pkl")
```
